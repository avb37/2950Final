```python
def count_vowel(str):
    """ takes string"""
    count = 0 
    vowel = set("aeiouyAEIOUY")
    for i in str:
        if i in vowel:
            count += 1
    return count
            
            
```

```python
def col_it(df, col1, col2, function):
    """ takes dataframe, column to iterate over, column to append, and chosen functio """
    temp = 0
    array = []
    for i in df[col1]:
        temp = function(i)
        array.append(temp)
    df[col2] = array
```

```python
def count_cons(str):
    """ takes string"""
    count = 0 
    vowel = set("bcdfghjklmnpqrstvwxzBCDFGHJKLMNPQRSTVWXZ")
    for i in str:
        if i in vowel:
            count += 1
    return count
            
            
```
